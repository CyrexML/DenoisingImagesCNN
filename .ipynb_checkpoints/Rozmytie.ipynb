```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import shutil
```

```python
# Путь к исходной папке с изображениями
source_folder = "D:/labSem6/images_dataset/ObrazyCzyste"
# Путь к папке для сохранения результирующих изображений с розмытием
blurred_folder = "D:/labSem6/images_dataset/BlurredImages"
# Путь к папке для сохранения изображений исходников
source_backup_folder = "D:/labSem6/images_dataset/SourceBackup"

# Создание папок, если они не существуют
os.makedirs(blurred_folder, exist_ok=True)
os.makedirs(source_backup_folder, exist_ok=True)

# Получаем список файлов в исходной папке
image_files = os.listdir(source_folder)
# Определяем количество изображений для розмытия (половину от всех)
num_images_to_blur = len(image_files) // 2

# Проходимся по каждому изображению и применяем розмытие Гаусса
for i in range(num_images_to_blur):
    image_path = os.path.join(source_folder, image_files[i])
    img = cv2.imread(image_path)
    blurred_img = cv2.GaussianBlur(img, (15, 15), 0)  # Пример размера ядра (15, 15), можно изменить
    # Сохраняем результирующее изображение в папке для розмытых изображений
    cv2.imwrite(os.path.join(blurred_folder, f"blurred_{image_files[i]}"), blurred_img)
    # Копируем исходное изображение в папку для резервного копирования
    shutil.copyfile(image_path, os.path.join(source_backup_folder, image_files[i]))
    # Удаляем исходное изображение из исходной папки
    os.remove(image_path)
```

```python
# Путь к исходной папке с изображениями
source_backup_folder = "D:/labSem6/images_dataset/SourceBackup"
# Путь к папке для сохранения результирующих изображений с розмытием
blurred_folder = "D:/labSem6/images_dataset/BlurredImages"

# Wybór kilku przykładowych obrazów do wyświetlenia
sample_images = np.random.choice(os.listdir(source_backup_folder), size=5, replace=False)

# Wyświetlanie obrazów przed и po zaszumieniu
fig, axes = plt.subplots(nrows=2, ncols=len(sample_images), figsize=(12, 6))
for i, image_file in enumerate(sample_images):
    clean_image = cv2.imread(os.path.join(source_backup_folder, image_file))
    blurred_image = cv2.imread(os.path.join(blurred_folder, f"blurred_{image_file}"))

    axes[0, i].imshow(cv2.cvtColor(clean_image, cv2.COLOR_BGR2RGB))
    axes[0, i].set_title("Czyste")
    axes[0, i].axis("off")

    axes[1, i].imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
    axes[1, i].set_title("Rozmyte")
    axes[1, i].axis("off")

plt.tight_layout()
plt.show()
```
