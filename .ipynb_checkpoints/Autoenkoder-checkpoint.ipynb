```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

plt.imshow(x_train[1], cmap="gray")
```

```python
print(x_train[0].shape)
```

```python
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
encoder_input = keras.Input(shape=(28,28,1), name="img")
x = keras.layers.Flatten()(encoder_input)
encoder_out = keras.layers.Dense(32, activation="relu")(x)

encoder = keras.Model(encoder_input, encoder_out, name="encoder")



decoder_input = keras.layers.Dense(784, activation="relu")(encoder_out)
# x = keras.layers.Dense(784, activation="relu")(decoder_input)
decoder_output = keras.layers.Reshape((28,28,1))(decoder_input)

opt = keras.optimizers.legacy.Adam(lr=0.001, decay = 1e-6)

autoencoder = keras.Model(encoder_input, decoder_output, name="autoencoder")
autoencoder.summary()

autoencoder.compile(opt, loss="mse")
autoencoder.fit(x_train, x_train, epochs=3, batch_size=32, validation_split=0.1)
```

```python
import random

def add_noise(img, random_chance = 5):
    noisy = []
    for row in img:
        new_row = []
        for pix in row:
            if random.choice(range(100)) <= random_chance:
                new_val = random.uniform(0,1)
                new_row.append(new_val)
            else:
                new_row.append(pix)
        noisy.append(new_row)
    return np.array(noisy)
```

Importing my dataset

```python
import tensorflow as tf
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator

# Paths to train and test
traindir = "D:/labSem6/ObrazyTrain"
testdir = "D:/labSem6/Obrazy"

datagen = ImageDataGenerator(rescale=1./255)

# Load train dataset
train_dataset = datagen.flow_from_directory(
    traindir,
    target_size=(256,256),
    batch_size=32,
    class_mode=None
)

# Load test dataset
test_dataset = datagen.flow_from_directory(
    testdir,
    target_size=(256,256),
    batch_size = 32,
    class_mode=None

)
```

```python
Building autoencoder model
```

```python
input_shape = (256, 256, 3)

# Build encoder
encoder = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters=16, kernel_size=3, strides=2, activation='relu', input_shape=input_shape),
    tf.keras.layers.Conv2D(filters=32, kernel_size=3, strides=2, activation='relu'),
    tf.keras.layers.Conv2D(filters=64, kernel_size=3, strides=2, activation='relu'),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=256, activation='relu'),
    tf.keras.layers.Dense(units=128, activation='relu')
])

# Build decoder
decoder = tf.keras.Sequential([
    tf.keras.layers.Dense(units=256, activation='relu', input_shape=(128,)),
    tf.keras.layers.Dense(units=4096, activation='relu'),
    tf.keras.layers.Reshape(target_shape=(16, 16, 16)),
    tf.keras.layers.Conv2DTranspose(filters=32, kernel_size=3, strides=2, activation='relu', padding='same'),
    tf.keras.layers.Conv2DTranspose(filters=16, kernel_size=3, strides=2, activation='relu', padding='same'),
    tf.keras.layers.Conv2DTranspose(filters=3, kernel_size=3, strides=2, activation='sigmoid', padding='same')
])

# Build autoencoder
autoencoder = tf.keras.Sequential([encoder, decoder])
```

Compiling and training model

```python
# Compile model
autoencoder.compile(optimizer='adam', loss='mse')

# Train model
history = autoencoder.fit(
    train_dataset,
    epochs=10,
    validation_data=test_dataset
)
```
